```python
import cv2
from ultralytics import YOLO

# Load trained YOLO model
model = YOLO("runs/detect/train/weights/best.pt")
cap = cv2.VideoCapture("Screen Recording 2026-01-15 173814.mp4")

# Open webcam (0 = laptop camera)
# cap = cv2.VideoCapture(0)

# For mobile camera (uncomment & fix URL if needed)
# cap = cv2.VideoCapture("http://192.168.29.227:4747/video")

if not cap.isOpened():
    print("❌ Camera not accessible")
    exit()

print("✅ Press Q to exit")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # YOLO inference
    results = model(frame, conf=0.13, verbose=False)

    # Draw detections
    for r in results:
        if r.boxes is not None:
            for box in r.boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                conf = float(box.conf[0])

                label = f"CRACK {conf:.2f}"

                # Draw box
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)

                # Draw label
                cv2.putText(
                    frame,
                    label,
                    (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.6,
                    (0, 0, 255),
                    2
                )

    # ✅ SHOW FRAME (FIXED)
    cv2.imshow("Real-Time Crack Detection", frame)

    # ✅ EXIT KEY (FIXED)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# ✅ CLEANUP (IMPORTANT)
cap.release()
cv2.destroyAllWindows()
```
